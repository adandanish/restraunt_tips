# **QUESTION (BASED ON FILTERING + VALUE COUNTS)**

# **Q2: For Speeding, were Men or Women stopped more often **

```python
# speeding is related to"violation' column
# men and women are from "driver_gender" column
data[data.violation == 'Speeding']
```

```python
# filtering command (to find specific value from a column)
# value counts (to find the sum of whole column )
# df[df.column_1 =='element/value'].column_2.value_counts()
data[data.violation =='Speeding']
```

```python
# this is telling us speeding of violation column
# by using whole command above
# df[df.column_1 =='element/value'].column_2.value_counts()
data[data.violation =='Speeding'].driver_gender.value_counts()
```

# **QUESTION (GROUPBY)**

# **Q3: does gender affect who get searched during a stop ?**

```python
# gender will be taken from "driver_gender"
# searched will be taken from 'search_conducted' column
# groupby method by using 2 columns
data.groupby('driver_gender').search_conducted.sum()
```

```python
# we want to know how much times search was conducted
data.search_conducted.value_counts()
```

# **so it is 2479 times**

# **QUESTION (MAPPING + DATA_TYPE CASTING)**

# **Q4: what is the mean stop_duration? **

```python
# by directly using mean command it will give us error
data.stop_duration.mean()
```

# as data in 'stop_duration' column is in "object" form i.e STRING TYPE
# so we will have to convert it into INTEGER
### **THIS** PROCESS WILL 'CALL DATA TYPE CASTING'

# **PROCESS OF MAPPING**

```python
data.stop_duration.value_counts()
```

```python
# it is telling us data is in 3 forms lets convert
# df['column name'] = df['column name'].map({old:new, old:new})
data['stop_duration'] = data['stop_duration'].map({'0-15 Min': 7.5, '16-30 Min' :24, '30+ Min': 45})
```
